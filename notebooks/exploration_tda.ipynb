```python
from gtda.time_series import takens_embedding_optimal_parameters
from tqdm_joblib import tqdm_joblib
from joblib import Parallel, delayed
from gtda.diagrams import Filtering
import numpy as np
from tqdm import tqdm
def find_best_embedding_mean(
    df: pd.DataFrame,
    serie_cols: list,
    max_time_delay: int,
    max_dimension: int,
    stride: int = 1,
    n_jobs: int = -1
) -> pd.DataFrame:
    """
    Encuentra los parmetros Takens ptimos para cada serie, luego calcula el promedio global.
    
    Retorna un DataFrame con los promedios y desviaciones estndar de time_delay y dimension.
    """

    X = df[serie_cols].to_numpy()

    def process_column(col_idx: int):
        ts = X[:, col_idx].reshape(-1, 1)
        try:
            time_delay, dimension = takens_embedding_optimal_parameters(
                ts,
                max_time_delay=max_time_delay,
                max_dimension=max_dimension,
                stride=stride,
                n_jobs=1,          # evitar paralelismo anidado
                validate=True
            )
        except Exception:
            time_delay, dimension = np.nan, np.nan
        return time_delay, dimension

    # Progreso paralelo
    with tqdm_joblib(tqdm(desc="Buscando parmetros Takens", total=len(serie_cols))):
        results = Parallel(n_jobs=n_jobs)(
            delayed(process_column)(col_idx) for col_idx in range(X.shape[1])
        )

    # Convertir a DataFrame
    results_df = pd.DataFrame(results, columns=["time_delay", "dimension"])
    valid = results_df.dropna()

    # Calcular promedio y desviacin
    summary = pd.DataFrame({
        "time_delay_mean": [valid["time_delay"].mean()],
        "time_delay_std": [valid["time_delay"].std()],
        "dimension_mean": [valid["dimension"].mean()],
        "dimension_std": [valid["dimension"].std()],
        "n_valid": [len(valid)],
        "n_total": [len(results_df)]
    })

    return summary
```

```python
from gtda.time_series import TakensEmbedding
from gtda.homology import VietorisRipsPersistence
from gtda.diagrams import Filtering, Scaler
from gtda.pipeline import Pipeline
from gtda.metaestimators import CollectionTransformer
from sklearn.decomposition import PCA

# ------------------ Parmetros ------------------
time_delay = 64
dimension = 8
stride = 3
n_components = 3  # por ejemplo, nmero de componentes PCA
homology_dimensions = [0, 1]  # H0 y H1

print(f"Imprimiendo optimos: {find_best_embedding_mean(df_prep_prin, df_prep_prin.columns, 100, 50, 3)}")
# ------------------ Construccin del pipeline ------------------
embedder = TakensEmbedding(
    time_delay=time_delay,
    dimension=dimension,
    stride=stride
)

batch_pca = CollectionTransformer(PCA(n_components=n_components, random_state=42), n_jobs=-1)
persistence = VietorisRipsPersistence(homology_dimensions=homology_dimensions, n_jobs=-1)
scaling = Scaler(n_jobs=-1)
filtering = Filtering(epsilon=0.075)  # 7.5% aprox se quitan

topo_pipeline = Pipeline([
    ("embedder", embedder),
    ("pca", batch_pca),
    ("persistence", persistence),
    ("scaling", scaling),
    ("filtering", filtering)
])

# ------------------ Aplicar pipeline ------------------
# df_prep_in debe ser un DataFrame o numpy array de shape (n_samples, n_timestamps)
X_transformed = topo_pipeline.fit_transform(df_prep_prin.values)

print(f"X_transformed: shape = {X_transformed.shape}")
```
