```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet('flask_app/results/confianza_matrix.parquet')
```

```python
# calcular la media por columna
means = df.mean(axis=0, skipna=True)

# estadísticos rápidos
print(means.describe())

# gráfica: distribución (hist + KDE) y boxplot
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(means, kde=True, bins=30, color='C0')
plt.xlabel('Mean value')
plt.title('Distribution of column means')

plt.subplot(1,2,2)
sns.boxplot(x=means, color='C1')
plt.xlabel('Mean value')
plt.title('Boxplot of column means')

plt.tight_layout()
plt.show()
```
