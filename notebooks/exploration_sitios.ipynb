```python
import pandas as pd
import os

# Carpeta con los CSVs de cierre agrícola
cierre_dir = "Sitios"

# Lista todos los CSVs que comienzan con "Cierre_agricola_mun"
csv_files = [f for f in os.listdir(cierre_dir) if f.startswith("Cierre_agricola_mun") and f.endswith(".csv")]

# Lista para ir guardando los DataFrames
dfs = []

for file in csv_files:
    file_path = os.path.join(cierre_dir, file)
    df = pd.read_csv(file_path, encoding='latin1')  # usa 'latin1' si da error con utf-8
    
    # Filtrar solo filas desde 2014
    df = df[df['Anio'] >= 2014]
    
    dfs.append(df)

# Concatenar todos los DataFrames
df_cierre_total = pd.concat(dfs, ignore_index=True)
df_cierre_total = df_cierre_total.sort_values(by='Anio').reset_index(drop=True)


# Guardar el CSV final
output_path = os.path.join(cierre_dir, "Cierre_agricola_total_2014_2024.csv")
df_cierre_total.to_csv(output_path, index=False, encoding='latin1')

print(f"✅ CSV guardado en: {output_path}")
print(f"Total filas después de concatenar: {len(df_cierre_total)}")
print(df_cierre_total['Anio'].min(), df_cierre_total['Anio'].max())
```
